```python
#Bubble sort
```

```python
list_to_sort = [3, 1, 9, 12, 4, 2]
```

```python
def bubble_sorting(list_to_sort):
    for n in range(len(list_to_sort) - 1):          # tyle przejść
        for i in range(len(list_to_sort) - 1):      # sąsiedzi: i oraz i+1
            if list_to_sort[i] > list_to_sort[i+1]:
                list_to_sort[i], list_to_sort[i+1] = list_to_sort[i+1], list_to_sort[i]
        print(list_to_sort)  
```

```python
bubble_sorting (list_to_sort)
```

```python
#Merge sort
```

```python
def merge(left, right):
    result = []
    i = 0
    j = 0

    # scalanie dwóch POSORTOWANYCH list
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    # jeśli coś zostało w jednej z list – dopisz
    result.extend(left[i:])
    result.extend(right[j:])

    return result
```

```python
def merge_sorting(list_to_sort):
    # PRZYPADEK BAZOWY
    # lista 0- lub 1-elementowa jest już posortowana
    if len(list_to_sort) <= 1:
        return list_to_sort

    # DZIELENIE
    srodek = len(list_to_sort) // 2
    left = list_to_sort[:srodek]
    right = list_to_sort[srodek:]

    # ODDANIE POŁÓWEK TEJ SAMEJ FUNKCJI
    left_sorted = merge_sorting(left)
    right_sorted = merge_sorting(right)

    # SCALANIE (tu się faktycznie sortuje)
    return merge(left_sorted, right_sorted)
```

```python
merge_sorting(list_to_sort)
```
